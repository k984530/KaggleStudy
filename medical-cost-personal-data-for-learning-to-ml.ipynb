```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# 데이터 불러오기

```python
data = pd.read_csv('../input/insurance/insurance.csv')
```

```python
data.head()
```

```python
X = np.c_[data['bmi']]
y = np.c_[data['charges']]
```

## LinearRegression

```python
import sklearn.linear_model
model = sklearn.linear_model.LinearRegression()
model.fit(X, y)
X_new = [[23]]
print(model.predict(X_new))
```

```python
a = np.c_[np.arange(15, 55, 0.1)]
b = model.predict(a)
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot() 
ax.plot(a, b, c='r')
ax.scatter(data.bmi, data.charges, c='g', s = 1, alpha=0.7)
plt.show()
```

## KNeighborRegressor

```python
import sklearn.neighbors
model = sklearn.neighbors.KNeighborsRegressor(n_neighbors=20)
model.fit(X, y)
X_new = [[23]]
print(model.predict(X_new))
```

```python
for i in range(1, 10, 3):
    model = sklearn.neighbors.KNeighborsRegressor(n_neighbors=i)
    model.fit(X, y)
    X_new = [[23]]
    print(model.predict(X_new))
    a = np.c_[np.arange(15, 55, 0.1)]
    b = model.predict(a)
    fig = plt.figure(figsize=(7,5))
    ax = fig.add_subplot() 
    ax.plot(a, b, c='r')
    ax.scatter(data.bmi, data.charges, c='g', s = 2, alpha=0.7)
    plt.show()
```
